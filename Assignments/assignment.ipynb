# Joining 2 tables 

```python
import pandas as pd
df1 = pd.DataFrame({'ID': [1, 2, 3],'Name': ['Ali', 'Amar', 'Charles']})
df2 = pd.DataFrame({'ID': [1, 2, 4],'Score': [85, 90, 78]})
print(df1)
print(df2)
```

```python
merged_df = df1.merge(df2, on='ID', how='inner')
```

```python
print(merged_df)
```

```python
merged_df = df1.merge(df2, on='ID', how='right')
print(merged_df)
```

```python
merged_df = df1.merge(df2, on='ID', how='left')
print(merged_df)
```
