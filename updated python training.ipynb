# HCF Of 2 numbers

```python

def hcf(num1,num2):
    if (num1<num2):
     small= num1
    else:
        small=num2
    for i in range(1,small+1):
        if((num1%i==0) and (num2%i==0)):
           hcf=i;
    return hcf
hcf(10,15)
```

# Merging 2 tables using merge()


```python
df1 = pd.DataFrame({'ID': [1, 2, 3],'Name': ['Ali', 'Amar', 'Charles']})
df2 = pd.DataFrame({'ID': [1, 2, 4],'Score': [85, 90, 78]})
print(df1)
print(df2)
```

```python
merged_df = df1.merge(df2, on='ID', how='inner')
```

```python
print(merged_df)
```

```python
merged_df = df1.merge(df2, on='ID', how='right')
print(merged_df)
```

```python
merged_df = df1.merge(df2, on='ID', how='left')
print(merged_df)
```
